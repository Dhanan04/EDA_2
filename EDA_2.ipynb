```python
#Ans1.)

import pandas as pd

# Load the dataset
flight_price_df = pd.read_excel('flight_price.xlsx')

# Examine the dimensions of the dataset
rows, columns = flight_price_df.shape
print("Number of rows:", rows)
print("Number of columns:", columns)
```

```python
#Ans2.) 

import matplotlib.pyplot as plt

# Create a histogram to visualize the distribution of flight prices
plt.figure(figsize=(10, 6))
plt.hist(flight_price_df['Price'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Flight Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
#Ans3.) 

# Calculate the minimum and maximum prices
min_price = flight_price_df['Price'].min()
max_price = flight_price_df['Price'].max()
price_range = max_price - min_price

print("Minimum Price:", min_price)
print("Maximum Price:", max_price)
print("Price Range:", price_range)
```

```python
#Ans4.)

# Create a boxplot to compare the prices of different airlines
plt.figure(figsize=(12, 8))
plt.boxplot([flight_price_df[flight_price_df['Airline'] == airline]['Price'] for airline in flight_price_df['Airline'].unique()],
            labels=flight_price_df['Airline'].unique())
plt.title('Flight Prices by Airline')
plt.xlabel('Airline')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
#Ans5.)

# Identify potential outliers using a boxplot
plt.figure(figsize=(10, 6))
plt.boxplot(flight_price_df['Price'], vert=False)
plt.title('Boxplot of Flight Prices')
plt.xlabel('Price')
plt.grid(True)
plt.show()
```

```python
#Ans9.) 

playstore_df = pd.read_csv('https://raw.githubusercontent.com/krishnaik06/playstore-Dataset/main/googleplaystore.csv')
print("Dimensions of the dataset:", playstore_df.shape)
```

```python
#Ans10.)

plt.figure(figsize=(12, 6))
boxplot = playstore_df.boxplot(column=['Rating'], by='Category', rot=90)
plt.title("Rating Variation by Category")
plt.ylabel("Rating")
plt.xlabel("Category")
plt.xticks(rotation=90)
plt.show()
```

```python
#Ans11.) 

missing_values = playstore_df.isnull().sum()
print("Missing values:\n", missing_values)
```

```python
#Ans12.) 

plt.figure(figsize=(8, 6))
plt.scatter(playstore_df['Size'], playstore_df['Rating'], alpha=0.5)
plt.title("Relationship between App Size and Rating")
plt.xlabel("Size")
plt.ylabel("Rating")
plt.show()
```

```python
#Ans13.) 

import pandas as pd

# Remove non-numeric characters from 'Price' column
playstore_df['Price'] = playstore_df['Price'].str.replace('$', '')

# Convert 'Price' column to numeric format
playstore_df['Price'] = pd.to_numeric(playstore_df['Price'], errors='coerce')

# Create a bar chart to compare average prices by app type
import matplotlib.pyplot as plt

avg_price_by_type = playstore_df.groupby('Type')['Price'].mean()

plt.bar(avg_price_by_type.index, avg_price_by_type.values)
plt.xlabel('App Type')
plt.ylabel('Average Price ($)')
plt.title('Average Price by App Type')
plt.show()
```

```python
#Ans14.)
# Remove rows with non-numeric 'Installs' values
playstore_df = playstore_df[playstore_df['Installs'].str.isnumeric()]

# Remove non-numeric characters from 'Installs' column
playstore_df['Installs'] = playstore_df['Installs'].str.replace(',', '').str.replace('+', '')

# Convert 'Installs' column to numeric format
playstore_df['Installs'] = pd.to_numeric(playstore_df['Installs'], errors='coerce')

# Remove rows with NaN values in 'Installs' column
playstore_df.dropna(subset=['Installs'], inplace=True)

# Identify the top 10 most popular apps
top_apps = playstore_df.nlargest(10, 'Installs')[['App', 'Installs']]
print("Top 10 most popular apps:\n", top_apps)
```

```python
#Ans15.) 

category_counts = playstore_df['Category'].value_counts().head(10)
category_counts.plot(kind='bar', figsize=(10, 6), color='green')
plt.title("Top 10 App Categories by Count")
plt.xlabel("Category")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```
